```python
import pandas as pd
```

```python
file=pd.read_csv('/content/Iris.csv')
file.columns
```

```python
x=file[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=file['Species']
```

```python
from sklearn.linear_model import Perceptron
model=Perceptron()
```

```python
model=model.fit(x,y)
```

```python
op=model.predict([[5,3,1,0.2]])
print(op)
```
